## Slide 30

```python
def kCDF(x, res=100, kernel='gau', bw=1):
    
    kde = sm.nonparametric.KDEUnivariate(x)
    kde.fit(kernel=kernel, bw=bw)
    interpdens = interp1d(kde.support, kde.density)
    xseq = np.linspace(min(kde.support), max(kde.support), res)
    
    holdout = np.zeros(res)
    for i in range(res):
        holdout[i] = integrate.quad(interpdens, min(kde.support), xseq[i])[0]
        
    CDF = interp1d(xseq, holdout)
    invCDF = interp1d(holdout, xseq)
    
    return CDF, invCDF
```
